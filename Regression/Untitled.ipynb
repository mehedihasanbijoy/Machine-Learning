```python
import numpy as np

X = np.array([1, 3, 5, 2, 4])
y = np.array([25, 35, 70, 35, 50])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.show()
```

```python
gradient_m = []
gradient_b = []
```

```python
m, b, learning_rate, number_of_iterations, n = 1, 1, .01, 1000, len(X)

for _ in range(number_of_iterations):
    ddm = -1/n * sum((y-m*X-b)*X)
    ddb = -1/n * sum((y-m*X-b))
    m = m - learning_rate * ddm
    b = b - learning_rate * ddb
    gradient_m.append(m)
    gradient_b.append(b)
```

```python
m, b
```

```python
plt.scatter(X, y)
plt.plot(X, [m*x + b for x in X], color = 'red')
plt.show()
```

```python
plt.scatter(gradient_m, gradient_b)
```
