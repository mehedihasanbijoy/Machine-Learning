```python
import pandas as pd

# loading the dataset
dataset = pd.read_csv('Datasets/Salary_Data.csv')
dataset.sample(5)
```

```python
# set(dataset['YearsExperience']), set(dataset['Salary'])
```

```python
from sklearn.model_selection import train_test_split

# declaring independent and dependent variable x and y
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# splitting data into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 32)
```

```python
from sklearn.linear_model import LinearRegression

# creating an instance of LinearRegression class and fitting data into it
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# prediction on test data
y_pred = regressor.predict(X_test)
```

```python
# showing original and predicted data
pd.DataFrame(list(zip(y_test, y_pred)), columns = ['y_test', 'y_pred'])
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

# visualizing training set data and the best fitted line
plt.scatter(X_train, y_train, color = 'k')
plt.plot(X_train, regressor.predict(X_train), color = 'r')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience [Training Set]')
plt.show()
```

```python
# visualizing test set data and the best fitted line
plt.scatter(X_test, y_test, color = 'purple')
plt.plot(X_train, regressor.predict(X_train), color = 'red')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience [Test Set]')
plt.show()
```
